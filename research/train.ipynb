```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load trained model
model = YOLO("food_segmentation_model_2.pt")

# Inference
results = model("nasi2.png", save=True, conf=0.3)
#show pixel sizes


# Show masks and classes
# for result in results:
#     result.show()  # shows image with boxes + masks
# Extract masks (binary masks per object)
masks = results[0].masks.data  # shape: (num_objects, H, W), type: torch.Tensor

# Calculate pixel area per mask
mask_areas = masks.sum(dim=(1, 2)).cpu().numpy()  # number of pixels per mask

# Output result
for i, area in enumerate(mask_areas):
    print(f"Object {i}: Mask pixel area = {int(area)} pixels")
    # print class
annotated_frame = results[0].plot()

# change from rgb to bgr
annotated_frame = cv2.cvtColor(annotated_frame, cv2.COLOR_RGB2BGR)

plt.imshow(annotated_frame)
plt.axis('off')
plt.show()
```

## Coin model training

```python
# train using yolo segment
import os
import sys
import ultralytics
from ultralytics import YOLO

os.environ["CUDA_LAUNCH_BLOCKING"] = "1"  # for Windows

# Load YOLOv8-segmentation model
model = YOLO("yolo11m-seg.pt")

# Train
model.train(
    data="coin_train/data.yaml",
    epochs=20,
    imgsz=640,
    batch=16,
    task="segment",
    name="coin_model"
)

# Save the model
model.save("coin_model.pt")
```

# Coin prediction and segmentation

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load trained model
model = YOLO("coin_model.pt")

# Inference
results = model("image2.png", save=True, conf=0.9)
#show pixel sizes

if results[0].masks is not None:
    masks = results[0].masks.data  # shape: (num_objects, H, W), type: torch.Tensor
    # Calculate pixel area per mask
    mask_areas = masks.sum(dim=(1, 2)).cpu().numpy()  # number of pixels per mask
    # Output result
    for i, area in enumerate(mask_areas):
        print(f"Object {i}: Mask pixel area = {int(area)} pixels")
        # print class
    annotated_frame = results[0].plot()

    # change from rgb to bgr
    annotated_frame = cv2.cvtColor(annotated_frame, cv2.COLOR_RGB2BGR)

    plt.imshow(annotated_frame)
    plt.axis('off')
    plt.show()
else:
    print("No masks found in the results.")
    plt.imshow(cv2.cvtColor(results[0].plot(),cv2.COLOR_RGB2BGR))
    plt.axis('off')
    plt.show()
```
